```python
map_input = {'A': 'Rock', 'B': 'Paper', 'C': 'Scissors', 'X': 'Rock', 'Y': 'Paper', 'Z': 'Scissors'}
points_shape = {'Rock': 1, 'Paper': 2, 'Scissors': 3}
points_outcome = {'Lose': 0, 'Draw': 3, 'Win': 6}
```

```python
with open('day02.txt') as f:
    lines = f.readlines()
    games = [val.strip() for val in lines]
print(games)
```

```python
#PART 1

def points_per_round(round_string):
    opponent_shape = map_input[round_string[0]]
    our_shape = map_input[round_string[2]]

    if opponent_shape == our_shape:
        return points_outcome['Draw'] + points_shape[our_shape]
    elif (opponent_shape, our_shape) in [('Paper', 'Rock'), ('Rock', 'Scissors'), ('Scissors', 'Paper')]:
        return points_outcome['Lose'] + points_shape[our_shape]
    else:
        return points_outcome['Win'] + points_shape[our_shape]
```

```python
[points_per_round(round_string) for round_string in games]
```

```python
sum([points_per_round(round_string) for round_string in games])
```

```python
# PART 2

map_input = {'A': 'Rock', 'B': 'Paper', 'C': 'Scissors', 'X': 'Lose', 'Y': 'Draw', 'Z': 'Win'}
points_shape = {'Rock': 1, 'Paper': 2, 'Scissors': 3}
points_outcome = {'Lose': 0, 'Draw': 3, 'Win': 6}
```

```python
def points_per_round2(round_string):
    opponent_shape = map_input[round_string[0]]
    outcome = map_input[round_string[2]]
    
    if (opponent_shape, outcome) in [('Rock', 'Draw'), ('Paper', 'Lose'), ('Scissors', 'Win')]:
        return points_outcome[outcome] + points_shape['Rock']
    elif (opponent_shape, outcome) in [('Rock', 'Win'), ('Paper', 'Draw'), ('Scissors', 'Lose')]:
        return points_outcome[outcome] + points_shape['Paper']
    else:
        return points_outcome[outcome] + points_shape['Scissors']
```

```python
sum([points_per_round2(round_string) for round_string in games])
```
